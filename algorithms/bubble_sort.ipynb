# Bubble sort

```python
from random import randint
```

```python
def pairs(arr): return zip(enumerate(arr), enumerate(arr[1:], 1))
def idx(pair): return pair[0]
def val(pair): return pair[1]
```

```python
def is_sorted(arr):
    for a, b in pairs(arr):
        if val(a) > val(b):
            return False
    return True
```

```python
def shift(arr, pair):
    i = idx(pair)
    v = arr.pop(i)
    arr.insert(i + 1, v)
```

```python
def bubble_sort(arr):
    while not is_sorted(arr):
        for a, b in pairs(arr):
            if val(a) > val(b):
                shift(arr, a)
```

```python
s = [randint(0, 100) for _ in range(25)]
print(s)
assert not is_sorted(s)
bubble_sort(s)
assert is_sorted(s)
print(s)
```
